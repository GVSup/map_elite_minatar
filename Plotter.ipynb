```julia
ENV["GKS_ENCODING"]="utf-8"
using Plots
using Random
using Statistics
```

```julia

trial1=[2,3,4,5,6]
trial2=[2.1,2.9,4.2,5,6]
trial3=[2.2,3,3.8,5,6]

n_trials=3

BehaviourX=[-1,0,1,2,3]
dim_behaviour=length(BehaviourX)

fits = zeros(dim_behaviour, n_trials)

fits[:, 1]=trial1
fits[:, 2]=trial2
fits[:, 3]=trial3

μ = mean(fits, dims=2)
σ = std(fits, dims=2)
#plot(BehaviourX, μ, ribbon=σ, label="Fitness", legend=:outertopright, xlabel="Espace des comportements", ylabel="Fitness")
plot(BehaviourX, μ, ribbon=σ, xlabel="Espace des comportements", ylabel="Fitness")
```
